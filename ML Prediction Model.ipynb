```python
#Check for the outliers in the “amount_in_usd” column and replace the outliers with appropriate values, discussed in the sessions.
import numpy as np

Q1 = df['amount_in_usd'].quantile(0.25)
Q3 = df['amount_in_usd'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

# to check the no of outliners
len(df[(df['amount_in_usd']>upper_bound) | (df['amount_in_usd']<lower_bound)])







```

```python
median_value = df['amount_in_usd'].median()
df['amount_in_usd'] = np.where((df['amount_in_usd'] < lower_bound) | (df['amount_in_usd'] > upper_bound),median_value,df['amount_in_usd'])
df
```

```python
#hyperparameter of xgboost
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
import numpy as np

# Split the data 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)

# Create an XGBoost regressor object
xg = xgb.XGBRegressor()

# Define the hyperparameter grid to search
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'subsample': [0.5, 0.8, 1.0],
    'colsample_bytree': [0.5, 0.8, 1.0],
}

random_search = RandomizedSearchCV(estimator=xg, param_distributions=param_grid, cv=5, n_iter=10, scoring='neg_mean_squared_error', random_state=42)
random_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = random_search.best_params_
print("Best hyperparameters:", best_params)


# Evaluate the model with the best hyperparameters on the test set
best_model = random_search.best_estimator_
predicted = best_model.predict(X_test)

MSE = mean_squared_error(y_test,predicted)
MAE = mean_absolute_error(y_test,predicted)
R2 = r2_score(y_test,predicted)

print('MSE =',MSE)
print('MAE =',MAE)
print('R-Square =',R2)

 
```
